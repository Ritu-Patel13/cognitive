```python
# My paragraph (Topic: Technology)
paragraph = """
Technology has revolutionized the way we live and work. With smartphones, laptops, and smart homes, 
our daily tasks have become easier and more efficient. Artificial intelligence is transforming industries 
and automating repetitive jobs. Cloud computing allows data access from anywhere. I love exploring the latest gadgets 
and innovations in the tech world.
"""
```

```python
## Lowercase and remove punctuation

import string

clean_text = paragraph.lower().translate(str.maketrans('', '', string.punctuation))
print(clean_text)
```

```python
# Tokenize into Words and Sentences

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize

nltk.download('punkt')

word_tokens = word_tokenize(clean_text)
sentence_tokens = sent_tokenize(clean_text)

print("Words:", word_tokens)
print("Sentences:", sentence_tokens)
```

```python
# Remove Stopwords + Frequency Distribution

from nltk.corpus import stopwords
from nltk.probability import FreqDist

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
filtered_words = [w for w in word_tokens if w not in stop_words]

print("Filtered Words:", filtered_words)

# Frequency distribution
fdist = FreqDist(filtered_words)
fdist.plot(title="Word Frequency (Excluding Stopwords)")
```

```python
#Stemming and Lemmatization


from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

porter = PorterStemmer()
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

porter_stemmed = [porter.stem(word) for word in filtered_words]
lancaster_stemmed = [lancaster.stem(word) for word in filtered_words]
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Original:", filtered_words)
print("Porter Stemmer:", porter_stemmed)
print("Lancaster Stemmer:", lancaster_stemmed)
print("Lemmatized:", lemmatized)
```

```python
#Regex & Text Splitting

import re

# a. Words with more than 5 letters
long_words = re.findall(r'\b\w{6,}\b', paragraph)
# b. Numbers
numbers = re.findall(r'\d+', paragraph)
# c. Capitalized words
capital_words = re.findall(r'\b[A-Z][a-z]*\b', paragraph)

# d. Alphabetic words only
alpha_words = re.findall(r'\b[a-zA-Z]+\b', paragraph)

# e. Words starting with a vowel
vowel_words = re.findall(r'\b[aeiouAEIOU][a-zA-Z]*\b', paragraph)

print("Words >5 letters:", long_words)
print("Numbers:", numbers)
print("Capitalized Words:", capital_words)
print("Alphabetic Words:", alpha_words)
print("Words starting with vowel:", vowel_words)
```

```python
#Custom Tokenization and Regex Cleaning

def custom_tokenizer(text):
    # Keep contractions and hyphenated words
    tokens = re.findall(r"\b(?:\w+(?:-\w+)*|'[a-z]+|\d+(?:\.\d+)?)\b", text)
    return tokens

sample_text = """
My email is user@example.com. Visit our site at https://www.example.com or call 123-456-7890 or +91 9876543210.
This is a state-of-the-art device that isn't available offline. It costs 3.14 times more!
"""

# Custom tokenization
custom_tokens = custom_tokenizer(sample_text)
print("Custom Tokens:", custom_tokens)

# Replace email, URL, phone
email_cleaned = re.sub(r'\b[\w.-]+?@\w+?\.\w{2,4}\b', '<EMAIL>', sample_text)
url_cleaned = re.sub(r'https?://[^\s]+', '<URL>', email_cleaned)
phone_cleaned = re.sub(r'(\+91\s?\d{10}|\d{3}-\d{3}-\d{4})', '<PHONE>', url_cleaned)

print("Cleaned Text:", phone_cleaned)
```
