```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter the value for M: "))


x = np.linspace(-10, 10, 400)


y1 = M * x**2
y2 = M * np.sin(x)


plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='M*x^2', color='blue', linestyle='--')
plt.plot(x, y2, label='M*sin(x)', color='green', linestyle='-.')

plt.title('Function Plots: M*x^2 and M*sin(x)')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Score': [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)


plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='viridis')


for i in range(len(df)):
    bar_plot.text(i, df['Score'][i] + 1, df['Score'][i], ha='center')

plt.title('Scores by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Replace with your actual roll number
roll_number = 1234
np.random.seed(roll_number)

data = np.random.randn(50)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot of cumulative sum
axs[0, 0].plot(np.cumsum(data), color='purple')
axs[0, 0].set_title('Cumulative Sum')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Cumulative Value')
axs[0, 0].grid(True)

# Scatter plot with random noise
noise = np.random.randn(50)
axs[0, 1].scatter(data, noise, color='orange')
axs[0, 1].set_title('Scatter with Noise')
axs[0, 1].set_xlabel('Data')
axs[0, 1].set_ylabel('Noise')
axs[0, 1].grid(True)

# Hide the unused subplots
axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read dataset from GitHub (raw link)
url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)

# 1. Total profit line plot
plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x='month_number', y='total_profit', marker='o')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()

# 2. Multi-line plot of all product sales
plt.figure(figsize=(10, 6))
product_columns = df.columns[1:-1]  # all product columns excluding month and total_profit
for col in product_columns:
    plt.plot(df['month_number'], df[col], label=col)

plt.title('Product Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar chart for all features (average per product)
df_mean = df[product_columns].mean()
plt.figure(figsize=(10, 5))
sns.barplot(x=df_mean.index, y=df_mean.values, palette='coolwarm')
plt.title('Average Sales per Product')
plt.ylabel('Average Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
