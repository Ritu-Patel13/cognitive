```python
import numpy as np
gfg = np.matrix('[4, 1, 9;12, 3, 1; 4, 5, 6]')
totalsum = np.sum(gfg)
print("the total sum is:",totalsum)

rowsum = np.sum(gfg,axis = 1)
print("row wise sum is :", rowsum)

columnsum = np.sum(gfg , axis = 0)
print("column wise sum is:", columnsum)

```

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])
sortedarray = np.sort(array)
print("sorted array is:", sortedarray)

sorted_indices = np.argsort(array)
print("sorted array indices is:", sorted_indices)

smallestelement = np.sort(array)[:4]
print("4 smallest elements are:",smallestelement)

largestelement = np.sort(array)[-5:]
print("5 largest element are:",largestelement)


array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integers = array[array.astype(int)==array]
print("the interger elements are :",integers)

floats = array[array.astype(int)!=array]
print("the floats elements are :",floats)



```

```python

import numpy as np
x= np.linspace(-10,10,100)

pow = 2
y = np.power(x,pow)

import matplotlib.pyplot as plt
%matplotlib inline

plt.title("y = x^2")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.plot(x,y)
plt.show()





```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = x**2  # Y = x^2
plt.plot(x, y1, label='y = x^2', linestyle='-', color='blue')


plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Plot of y = x^2')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = np.sin(x)  # Y = sin(x)

plt.plot(x, y1, label='y = sin(x)', linestyle='-', color='blue')

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Plot of y = sin(x)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = np.exp(x)  # Y = e^x

plt.plot(x, y1, label='y = e^x', linestyle='-', color='blue')

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Plot of $y = e^x$')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = np.log(np.abs(x) + 1)  # Y = log(|x| + 1)

plt.plot(x, y1, label='y = e^x', linestyle='-', color='blue')

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Plot of y = log(|x| + 1)')
plt.legend()
plt.grid(True)

plt.show()
```
